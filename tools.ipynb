```python
def move_reaction_lines(source_file, target_file):
    """
    ä» A.txt ä¸­åˆ é™¤æ‰€æœ‰åŒ…å« 'Reaction:' çš„è¡Œï¼Œå¹¶å°†è¿™äº›è¡Œç§»åŠ¨åˆ° B.txtã€‚
    :param source_file: åŸå§‹æ–‡ä»¶ (A.txt)
    :param target_file: ç›®æ ‡æ–‡ä»¶ (B.txt)
    """
    try:
        with open(source_file, "r", encoding="utf-8") as f:
            lines = f.readlines()

        # åˆ†ç¦»åŒ…å« 'Reaction:' çš„è¡Œ å’Œ å…¶ä»–è¡Œ
        reaction_lines = [line for line in lines if "Reaction:" in line]
        remaining_lines = [line for line in lines if "Reaction:" not in line]

        # å°†å‰©ä½™å†…å®¹å†™å› A.txt
        with open(source_file, "w", encoding="utf-8") as f:
            f.writelines(remaining_lines)

        # å°† Reaction è¡Œè¿½åŠ åˆ° B.txt
        if reaction_lines:
            with open(target_file, "a", encoding="utf-8") as f:
                f.writelines(reaction_lines)

        print(f"âœ… å¤„ç†å®Œæˆï¼šå·²åˆ é™¤ A.txt ä¸­çš„ 'Reaction:' è¡Œï¼Œå¹¶è½¬ç§»åˆ° B.txtã€‚")

    except Exception as e:
        print(f"âŒ å¤„ç†æ–‡ä»¶æ—¶å‡ºé”™: {e}")


# è°ƒç”¨å‡½æ•°ï¼ŒæŒ‡å®š A.txt å’Œ B.txt çš„æ–‡ä»¶è·¯å¾„
move_reaction_lines("simulate.txt", "B.txt")
```

```python
import json

# è¾“å…¥å’Œè¾“å‡ºæ–‡ä»¶è·¯å¾„
input_txt_file = "250313.txt"  # ä½ çš„æ—¥å¿—æ–‡ä»¶è·¯å¾„
output_txt_file = "extracted_methods.txt"  # ç»“æœä¿å­˜æ–‡ä»¶ï¼ˆå¯é€‰ï¼‰

def extract_unique_methods(input_file, output_file=None):
    """
    ä»æ—¥å¿—æ–‡ä»¶ä¸­æå– `method` å­—æ®µï¼Œå¹¶å»é‡ã€æ’åºã€‚
    """
    extracted_methods = set()  # ä½¿ç”¨ `set` è‡ªåŠ¨å»é‡
    prefix = "LiqiMsg: "  # LiqiMsg: åçš„ JSON æ•°æ®

    with open(input_file, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line.startswith(prefix):  # è¯†åˆ« "LiqiMsg: " å¼€å¤´çš„è¡Œ
                json_str = line[len(prefix):]  # æå– JSON éƒ¨åˆ†
                try:
                    liqi_msg = json.loads(json_str)  # è§£æ JSON
                    method = liqi_msg.get("method")  # è·å– "method" å­—æ®µ
                    if method:
                        extracted_methods.add(method)  # æ·»åŠ åˆ° `set`ï¼Œè‡ªåŠ¨å»é‡
                except json.JSONDecodeError:
                    print(f"âŒ è§£æ JSON å¤±è´¥: {line}")

    # è½¬æ¢ä¸ºåˆ—è¡¨å¹¶æ’åº
    sorted_methods = sorted(extracted_methods)

    # ç»“æœæ‰“å°
    print("\nğŸ“Œ æå–çš„å”¯ä¸€ `method` å€¼ï¼ˆå·²å»é‡ï¼‰ï¼š")
    for method in sorted_methods:
        print(method)

    # å¦‚æœæä¾›äº†è¾“å‡ºæ–‡ä»¶è·¯å¾„ï¼Œåˆ™å°†ç»“æœå†™å…¥æ–‡ä»¶
    if output_file:
        with open(output_file, "w", encoding="utf-8") as out_f:
            for method in sorted_methods:
                out_f.write(method + "\n")
        print(f"\nâœ… æå–ç»“æœå·²ä¿å­˜åˆ° `{output_file}`")

# è¿è¡Œæå–
extract_unique_methods(input_txt_file, output_txt_file)
```
